# Esplorazione delle varietà di vino più apprezzate:

```python
# Top 10 varietà di vino acquistate più di frequente
top_varieties = df['variety'].value_counts().head(10)
print(top_varieties)

# Grafico a barre delle top varietà
plt.figure(figsize=(10, 6))
sns.barplot(x=top_varieties.values, y=top_varieties.index, palette='viridis')
plt.title('Top 10 Varietà di Vino più apprezzate')
plt.xlabel('Numero di Vini')
plt.ylabel('Varietà di Vino')
plt.show()
```

# Esplorazione delle vigne più apprezzate:

```python
# Top 10 produttori di vino più frequenti

top_wineries = df['winery'].value_counts().head(10)
print(top_wineries)

# Grafico a barre delle top vigne
plt.figure(figsize=(10, 6))
sns.barplot(x=top_wineries.values, y=top_wineries.index, palette='viridis')
plt.title('Top 10 Produttori di Vino più apprezzati')
plt.xlabel('Numero di Vini')
plt.ylabel('Produttore di Vino')
plt.show()
```

```python
# Top 10 titoli di vino acquistati più di frequente
top_titles = df['title'].value_counts().head(10)
print(top_titles)

# Grafico a barre dei top titoli
plt.figure(figsize=(10, 6))
sns.barplot(x=top_titles.values, y=top_titles.index, palette='viridis')
plt.title('Top 10 Titoli di Vino più apprezzati')
plt.xlabel('Numero di Vini')
plt.ylabel('Titoli di Vino')
plt.show()
```

# Esplorazione degli outlier di prezzo:

```python
# Analisi della distribuzione dei prezzi

price_distribution = df['price'].describe()
print(price_distribution)

# Boxplot dei prezzi
plt.figure(figsize=(10, 6))
sns.boxplot(x='price', data=df, palette='viridis')
plt.title('Distribuzione dei Prezzi dei Vini')
plt.xlabel('Prezzo')
plt.show()
```

```python
from scipy import stats
import numpy as np

# Calcola il punteggio Z per la colonna 'price'
z_scores = stats.zscore(df['price'])

# Imposta una soglia per identificare gli outlier (ad esempio, 3 deviazioni standard)
threshold = 3
outliers = np.abs(z_scores) > threshold

# Conta il numero di outlier
num_outliers = np.sum(outliers)

print(f"Numero di outliers nella colonna 'price': {num_outliers}")
```

# Esplorazione delle varietà più costose:

```python
# Top 10 varietà di vino più costose

top_varieties = df.groupby('variety')['price'].mean().sort_values(ascending=False).head(10)
print(top_varieties)
```

# Esplorazione dei titoli più costosi:

```python
# Top 10 titoli di vino più costosi

top_titles = df.groupby('title')['price'].mean().sort_values(ascending=False).head(10) 
print(top_titles)
```

# Esplorazione dei paesi con i vini più costosi:

```python
# Analisi di paesi con i vini più costosi
top_countries = df.groupby('country')['price'].mean().sort_values(ascending=False).head(10)
print(top_countries)
```

# Esplorazione delle variabili categoriche:

```python
# Boxplot del punteggio in base al paese
plt.figure(figsize=(12, 8))
sns.boxplot(x='country', y='points', data=df, palette='viridis')
plt.title('Distribuzione del Punteggio dei Vini per Paese')
plt.xlabel('Paese')
plt.ylabel('Punteggio')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Analisi della correlazione tra prezzo e punteggio:

```python
# Scatter plot tra punteggio e prezzo
plt.figure(figsize=(10, 6))
sns.scatterplot(x='points', y='price', data=df, color='darkred')
plt.title('Scatter Plot tra Punteggio e Prezzo dei Vini')
plt.xlabel('Punteggio')
plt.ylabel('Prezzo')
plt.show()

# Calcolo della correlazione tra punteggio e prezzo
correlation_score_price = df['points'].corr(df['price'])
print("Correlazione tra punteggio e prezzo:", correlation_score_price)
```

In questo caso, il valore di 0.416 suggerisce che c'è una tendenza positiva tra il punteggio e il prezzo dei vini. Ciò significa che, in generale, i vini con punteggi più alti tendono ad avere prezzi più alti. Tuttavia, la correlazione non implica causalità, e ci possono essere molteplici fattori che contribuiscono a questa relazione.
